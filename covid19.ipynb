```python
filtered_data = data[data['location'] == 'Kenya']
filtered_data
```

```python
data.fillna(0, inplace=True)
```

```python
data['date'] = pd.to_datetime(data['date'])

data.head()
```

```python
selected_countries = ['USA', 'Italy', 'India']
df_selected = data[data['location'].isin(selected_countries)]
location = 0
plt.figure(figsize=(10,6))
for country in selected_countries:
    subset = df_selected[df_selected['location'] == location]
    plt.plot(subset['date'], subset['total_cases'], label=location)

plt.title('Total Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ------------------------------------------
# 2. Plot total deaths over time for selected countries
plt.figure(figsize=(10,6))
for country in selected_countries:
    subset = df_selected[df_selected['location'] == location]
    plt.plot(subset['date'], subset['total_deaths'], label=location)

plt.title('Total Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ------------------------------------------
# 3. Compare daily new cases between countries
data['daily_new_cases'] = data.groupby('location')['total_cases'].diff().fillna(data['total_cases'])

plt.figure(figsize=(10,6))
for country in selected_countries:
    subset = data[data['location'] == location]
    plt.plot(subset['date'], subset['daily_new_cases'], label=location)

plt.title('Daily New Cases Comparison')
plt.xlabel('Date')
plt.ylabel('Daily New Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ------------------------------------------
# 4. Calculate death rate (total_deaths / total_cases)
data['death_rate'] = data['total_deaths'] / data['total_cases']
data['death_rate'] = data['death_rate'].fillna(0)

print(data[['date', 'location', 'total_cases', 'total_deaths', 'death_rate']].head(15))

# ------------------------------------------
# Additional Visualizations

# Bar chart: Top countries by total cases (latest date)
latest_date = data['date'].max()
latest_data = data[data['date'] == latest_date].sort_values('total_cases', ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x='location', y='total_cases', data=latest_data, palette='viridis')
plt.title(f'Top Countries by Total Cases on {latest_date.date()}')
plt.xlabel('location')
plt.ylabel('Total Cases')
plt.tight_layout()
plt.show()

# Heatmap: Correlation matrix of numeric columns
corr = data[['total_cases', 'total_deaths', 'daily_new_cases', 'death_rate']].corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
selected_countries = ['USA', 'Italy', 'India']
df_selected = data[data['location'].isin(selected_countries)]

plt.figure(figsize=(10,6))
for location in selected_countries:
    subset = df_selected[df_selected['location'] == location]
    plt.plot(subset['date'], subset['life_expectancy'], label=location)

plt.title('Cumulative Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Cumulative Vaccinations')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
